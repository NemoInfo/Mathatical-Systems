```python
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
plt.style.use('Solarize_Light2')
import numpy as np
```

## Part 1: Plotting MAE overtime

```python
f = open("data/part1_out.txt")
data = [[*map(float, line.strip().split(" "))] for line in f.readlines()]
data_def, data_alt = zip(*data)
```

```python
k = len(data)
xs = np.arange(0, k * 1000 + 1, 1000)
fig, ax = plt.subplots()

plt.plot(xs[1:], data_def, "b", linewidth=1, label="Benchmark SR")
plt.plot(xs[1:], data_alt, "r", linewidth=1, label="Alternative SR")

plt.legend()
plt.gcf().set_size_inches(8, 4)
plt.gcf().set_dpi(100)
plt.xticks(xs[::500])
ax.xaxis.set_major_formatter(FuncFormatter(lambda x, pos: f"{x // 1000}"))
ax.set_ylim(top=6e-10) # remove this for non-zommed in version
plt.xlabel("Iteration (1e3)")
plt.ylabel("Mean Absolute Error")

# plt.savefig('part1_zoom.svg') 
```

# Part 2: Plotting AE overtime

```python
f = open("data/part2_out.txt")
data = [[*map(float, line.strip().split(" "))] for line in f.readlines()]
data_rrn, data_rsr, data_csm = zip(*data)
```

```python
k = len(data)
xs = np.arange(1_000_000, k * 1_000_000 + 1, 1_000_000)
fig, ax = plt.subplots()

plt.plot(xs[:], data_rrn, "y", linewidth=1, label="recursive round-to-nearest")
plt.plot(xs[:], data_rsr, "r", linewidth=1, label="recursive stochastic rounding")
plt.plot(xs[:], data_csm, linewidth=1, label="compensated summation")

plt.legend(loc='center', bbox_to_anchor=(0.75, 0.77))
plt.gcf().set_size_inches(8, 4)
plt.gcf().set_dpi(100)
plt.xticks(xs)

ax.set_xscale('log')
ax.set_yscale('log')
plt.xlabel("Iteration")
plt.ylabel("Absolute Error")

# plt.savefig('part2.svg')
```
